```python
import datetime

date = datetime.date(2025, 1, 2) #year, month, day

print(date) #return 2025-01-02

today = datetime.date.today()
print(today) #should print the current date in format yyyy-mm-dd
```

```python
import datetime

time = datetime.time(12, 30, 0) #hrs, mins, seconds
print(time) #will return 12:30:00

now = datetime.datetime.now()
print(now) #returns the current date and time in format yyyy-mm-dd hr:min:sec.milisecond???

now = now.strftime("%H:%M:%S") #strftime is stringformat for time
print(now) #will return the current time in format hr:min:sec

now = now.strftime("%H:%M:%S %m-%d-%Y")
print(now) #will return the current time in format hr:min:sec and date in format mm-dd-yyy
```

```python
target_datetime = datetime.datetime(2030, 1, 2, 12, 30, 1)
current_datetime = datetime.datetime.now()

if target_datetime < current_datetime:
    print("Target date has passed")
else:
    print("Target date has NOT passed")
```
