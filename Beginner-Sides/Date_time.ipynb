```python
import datetime

date = datetime.date(2025, 1, 2) #year, month, day

print(date) #return 2025-01-02

today = datetime.date.today()
print(today) #should print the current date in format yyyy-mm-dd
```

```python
import datetime

time = datetime.time(12, 30, 0) #hrs, mins, seconds
print(time) #will return 12:30:00

now = datetime.datetime.now()
print(now) #returns the current date and time in format yyyy-mm-dd hr:min:sec.milisecond???

time_only = now.strftime("%H:%M:%S")
time_and_date = now.strftime("%H:%M:%S %m-%d-%Y")

print(time_only) #will print time in format hr:min:second
print(time_and_date) #will print date and time in format hr:min:sec mm-dd-yyyy
```

```python
import datetime

current_date = datetime.datetime.now()
print(current_date)

day = current_date.strftime("%A") #will print the current Day
print(day)

month = current_date.strftime("%B")
print(month)
```

```python
target_datetime = datetime.datetime(2030, 1, 2, 12, 30, 1) # target date in format: yyyy, m, d, hr, min, second
current_datetime = datetime.datetime.now()

if target_datetime < current_datetime:
    print("Target date has passed")
else:
    print("Target date has NOT passed")
```

```python
import time

print(time.ctime(0)) #this is the epoch or beginning of time for a computer - wed dec 31 1969 @ 1900 hrs in this case

#if you put a time (in seconds where 0 is that is the time it will be that many seconds from epoch)

print(time.ctime(631139040)) #time / date approx 20yrs from epoch

print(time.time()) #time since epoch in seconds to right nows time

print(time.ctime(time.time())) #time.time will return seconds since epoch and ctime makes it readable as date / time


time_obj = time.localtime()

print(time_obj) #a stucture containing year, month, day, hr, min, second, etc

#time.strftime(format, timeobj) ex.
local_time = time.strftime("%B %d %Y %H:%M:%S", time_obj) #will print time with month, day, year, hr:min:sec
print(local_time)
```
